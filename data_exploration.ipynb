```python
#import excel file
cycling_data_1 = pd.read_excel('Cycling_1.xls')
```

```python
cycling_data_2 = pd.read_excel('Cycling_2.xls')
cycling_data_3 = pd.read_excel('Cycling_3.xls')
metro_data_1 = pd.read_excel('Metro_1.xls')
metro_data_2 = pd.read_excel('Metro_2.xls')
walking_data_1 = pd.read_excel('Walking_1.xls')
walking_data_2 = pd.read_excel('Walking_2.xls')
walking_data_3 = pd.read_excel('Walking_Shopping.xls')
```

```python
metro_data_1
```

```python
#merg the three cycling data sets
cycling_data = pd.concat([cycling_data_1, cycling_data_2, cycling_data_3])
cycling_data
```

```python
#rename the columns
cycling_data 
```

```python
#rename the third column to Ay_cycling
cycling_data.rename(columns={'Acceleration y (m/s^2)':'Ay_cycling'}, inplace=True)
cycling_data
```

```python
cycling_data.rename(columns={'Acceleration x (m/s^2)':'Ax_cycling'}, inplace=True)
cycling_data
```

```python
cycling_data.rename(columns={'Acceleration z (m/s^2)':'Az_cycling'}, inplace=True)
cycling_data
```

```python
#merge the two metro data sets
metro_data = pd.concat([metro_data_1, metro_data_2])
metro_data
```

```python
metro_data.rename(columns={'Acceleration x (m/s^2)':'Ax_metro', 'Acceleration y (m/s^2)':'Ay_metro', 'Acceleration z (m/s^2)':'Az_metro', 'X (m/s^2)':'X_metro','Y (m/s^2':'Y_metro', 'Z (m/s^2)':'Z_metro'}, inplace=True)
metro_data
```

```python
metro_data.rename(columns={'Y (m/s^2)':'Y_metro'}, inplace=True)
```

```python
metro_data
```

```python
#merge the three walking data sets
walking_data = pd.concat([walking_data_1, walking_data_2, walking_data_3])
walking_data
```

```python
walking_data.rename(columns={'Acceleration x (m/s^2)':'Ax_walking', 'Acceleration y (m/s^2)':'Ay_walking', 'Acceleration z (m/s^2)':'Az_walking', 'X (m/s^2)':'X_walking', 'Y (m/s^2':'Y_walking','Z (m/s^2)':'Z_walking' }, inplace=True)
```

```python
walking_data
```

```python
walking_data.rename(columns={'Y (m/s^2)':'Y_walking'}, inplace=True)
```

```python
#merge the three data sets
data = pd.concat([cycling_data, metro_data, walking_data])
data
```
