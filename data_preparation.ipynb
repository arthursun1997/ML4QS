```python
window_size = 100 
columns_mean = ['cycling_A_x', 'cycling_Gyr_x', 'cycling_freq', 'cycling_speed', 'walking_A_x', 'walking_Gyr_x', 'walking_freq', 'walking_speed', 'metro_A_x', 'metro_Gyr_x', 'metro_frequ', 'metro_doppler_speed' ]  
```

```python
#slice dataframe into windows of 100 rows
df_mean = merged_all[columns_mean].rolling(window_size).mean().iloc[window_size - 1::window_size, :]
df_mean
```

```python
#get max values for each window
df_max = merged_all[columns_mean].rolling(window_size).max().iloc[window_size - 1::window_size, :]
df_max
```

```python
#get min values for each window
df_min = merged_all[columns_mean].rolling(window_size).min().iloc[window_size - 1::window_size, :]
df_min
```

```python
#get standard deviation for each window
df_std = merged_all[columns_mean].rolling(window_size).std().iloc[window_size - 1::window_size, :]
df_std
```

```python
transformed_df = pd.DataFrame()
```

```python
for col in columns_mean:
    column_data = merged_all[col].values
    num_windows = len(column_data) // 100
    
    transformed_col = []
    for i in range(num_windows):
        window = column_data[i*100 : (i+1)*100]
        window_transform = np.fft.fft(window)
        transformed_col.extend(window_transform)
    
    transformed_df[col] = transformed_col

# Print the resulting transformed DataFrame
print(transformed_df)
```

```python
column_to_transform = 'cycling_A_x'

# Get the data from the selected column
column_data = merged_all[column_to_transform].values

# Apply Fourier transformation
transformed_data = np.fft.fft(column_data)

# Calculate the frequencies corresponding to the transformed data
frequencies = np.fft.fftfreq(len(column_data))

# Plot the Fourier transformation
plt.plot(frequencies, np.abs(transformed_data))
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.title('Fourier Transformation of {}'.format(column_to_transform))
plt.show()
```
